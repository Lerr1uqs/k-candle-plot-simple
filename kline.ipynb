```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
df = pd.read_csv("assets/yiling.csv")
df
```

```python
# not use the "Adj Close" colomn
selected_colomns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
data = df[selected_colomns]
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)  # 直接设置索引，而不是创建一个新的列再删除原来的列
```

```python
data
```

```python
mpf.plot(
    data,
    type="candle",
    title="Yiling Pharmaceutical",
    style="yahoo",
    volume=True,
    figratio=(12.00, 8), # figratio modifies the aspect ratio of the plot.
    returnfig=True,
    show_nontrading=False # Non-trading periods can be displayed using the show_nontrading kwarg.
)
```
